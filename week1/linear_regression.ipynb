#### step 1: import package

```python
import pandas as pd
import numpy as np
```

#### step 2: load data

```python
df = pd.read_csv('train.csv')
df.head(18)
```

```python
df1 = df[df['observation'] == 'PM2.5']
df1.head()
```

```python
df2 = df1.iloc[:, 3: ]
df2.head()
```

**There are 24 features in one day, we need split it to 10 continuous features. 24-10+1 = 15**

```python
train_data = np.array(df2, float)
# normlize data
train_data = (train_data - np.mean(train_data)) / np.std(train_data)
print(train_data)
X_train = []
y_train = []
for i in range(15):
    X_train.extend(train_data[:, i: i + 9])
    y_train.extend(train_data[:, i + 9])
```

```python
X_train = np.array(X_train, float)
X_train = np.concatenate((np.ones((X_train.shape[0],1)), X_train),axis=1) # add bias
y_train = np.array(y_train, float)
print("train features X shape: ", X_train.shape)
print("train label y shape: ", y_train.shape)
```

#### step 3: train model

```python
iters = 100000
m, n = X_train.shape # dimension of features
W = np.random.randn(n)
losses = []
# X_train = (X_train - np.mean(X_train)) / np.std(X_train)
# y_train = (y_train - )
lr = 0.001 # learning rate
# print(X_train)
print(W)
for i in range(iters):
    y_pred = X_train.dot(W)
    loss = 1/ 2.0 * ((y_pred - y_train) ** 2).sum()
    losses.append(loss / m)
    W_grad = np.dot(X_train.transpose(), (y_pred - y_train)) 
    W -= lr * W_grad / float(m)
    if i % 10000 == 0:
        print("iters {} , learning rate: {}, loss: {}".format(i, lr, loss / float(m)))
    if iters == i * 2:
        lr *= 0.1
print(W)
```
